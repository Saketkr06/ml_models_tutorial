```python
import pandas as pd
import numpy as np
```

```python
teams=pd.read_csv("/content/teams.csv")
```

```python
teams
```

```python
x = teams[["athletes","prev_medals"]].copy()
y = teams[["medals"].copy()]
```

```python
y
```

```python
x["intercept"]=1
```

```python
x = x[["intercept","athletes","prev_medals"]]
```

```python
x
```

```python
x_T =x.T
```

```python
x_T
```

```python
B = np.linalg.inv(x_T @ x) @ x_T @ y
```

```python
B
```

```python
B.index = x.columns
```

```python
B
```

```python
predictions = x@ B
```

```python
predictions
```

```python
SSR = ((y - predictions) ** 2).sum()
```

```python
SST = ((y - y.mean()) ** 2).sum()
```

```python
SSR
```

```python
SST
```

```python
R2 = 1 - (SSR / SST)
```

```python
R2
```
