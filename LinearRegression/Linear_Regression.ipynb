### **Definition**
Linear Regression is a model in which the independent and dependent variables follows a linear relationship.



> y = mx + c , where y is a dependent variable and x is an independent variable.

### **Importing Libraries for Data Analysis and Visualisation**

```
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

The below .csv file contains marks of some students with their hours of study. Hence, this data follows a linear relationship as the no. of hours of study will increase then the marks should increase too. 

```
df = pd.read_csv('/content/Linear1.csv')
df.head()
```

```
plt.figure(figsize = (12,6))
plt.scatter(df['Hours'], df['Marks'])
plt.xlabel('Hours')
plt.ylabel('Marks')
plt.title('Single Variable LR')
plt.show()
```

## **Cost Function**
The cost is the difference between estimated values, or the hypothesis and the real values — the actual data we are trying to fit a line to.

Capture.PNG

### **Gradient Descent**
For finding the minima of the cost function we use gradient descent which step downs the cost function in the direction of the steepest descent.
The choice of correct learning rate is very important as it ensures that Gradient Descent converges in a reasonable time. :

### Importing and Using Linear Regression from scikit learn

```
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```
x = df.drop(['Marks'], axis = 1)
y = df['Marks']
```

```
model.fit(x,y)
```

```
model.predict([[2]])
```

```
model.coef_
```

```
model.intercept_
```

```
hours = [[3], [5], [10], [8], [12], [13], [2]]
model.predict(hours)
```

### **Linear Regression with Multiple Variables**

```
df2 = pd.read_csv('/content/Linear2.csv')
df2.head()
```

```
plt.figure(figsize = (12,6))
plt.scatter(df2['Study Hours'], df2['Marks'], color = 'red')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title('Multiple Variable LR')
plt.show()
```

```
plt.figure(figsize = (12,6))
plt.scatter(df2['Playing Hours'], df2['Marks'], color = 'green')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title('Multiple Variable LR')
plt.show()
```

```
x2 = df2.drop(['Marks'], axis = 1)
y2 = df2['Marks']
```

```
x2
```

```
model2 = LinearRegression()
model2.fit(x2, y2)
```

```
model2.predict([[7, 4, 3]])
```
